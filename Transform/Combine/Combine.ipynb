```python
# Import pandas library
import pandas as pd
  
# initialize list of lists
data = [['Sara', 10], ['Mohab', 15], ['Ahmed', 14], ['Misk' , 3]]
  
# Create the pandas DataFrame
df1 = pd.DataFrame(data, columns=['Name', 'Age'])
  
# print dataframe.
df1
```

```python
# initialize list of lists
data = [['Python', 10000], ['PHP', 7000], ['C++', 9000], ['JavaScript' , 10000]]
  
# Create the pandas DataFrame
df2 = pd.DataFrame(data, columns=['Job', 'Salary'])
  
# print dataframe.
df2
```

```python
# Combine tow dataset using (Concat) method .
df_combin = pd.concat([df1 , df2], axis=1) # Must be add axis , because non make missing values .
df_combin.head(2)
```

# Pandas Melt

 لجعل تحليل البيانات في الجدول أسهل ، يمكننا إعادة تشكيل البيانات في أشكال أكثر ملاءمة  باستخدام الدالة
 - Melt 
 

```python
# Name is id_vars and Course is value_vars
pd.melt(df_combin, id_vars =['Name'], value_vars =['Job'])
```

```python
# multiple  columns
pd.melt(df_combin, id_vars =['Name'], value_vars =['Job', 'Salary'])
```

```python
# Names of ‘variable’ and ‘value’ columns can be customized
pd.melt(df_combin, id_vars =['Name'], value_vars =['Job'],
              var_name ='Job Columne', value_name ='Job Labguage')
```
